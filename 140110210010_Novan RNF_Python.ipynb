# Elemen Ekuivalen Siklik

```python
# Mencari elemen-elemen yang ekuivalen siklik
def Mencari_Elemen_Ekuivalen_siklik(kodons):
    # Mencari permutasi dari kodon
    def permutasi_kodons(kodons):
        
        # Permutasi a_1 = (1 2 3)
        keys_1 = [2, 3, 1]
        # Permutasi a_2 = (1 3 2)
        keys_2 = [3, 1, 2]

        keys_11 = [key - 1 for key in keys_1]
        keys_22 = [key - 1 for key in keys_2]

        # Periksa apakah panjang keys sama dengan panjang kodons
        if len(keys_11) != len(kodons) and len(keys_22) != len(kodons):
            raise ValueError('Panjang keys tidak sama dengan panjang kodons')
        
        # Mengembalikan permutasi terhadap kodons_i
        kodons_1 = [kodons[key] for key in keys_11]
        kodons_2 = [kodons[key] for key in keys_22]
        return ''.join(kodons_1) if isinstance(kodons, str) else kodons_1, ''.join(kodons_2) if isinstance(kodons, str) else kodons_2
    
    # membentuk himpunan dari elemen-elemen yang ekuivalen siklik
    def elemen_ekuivalen_siklik(kodons):
        
        # Membentuk dictionary untuk menyimpan urutan kodon
        urutan_kodon = {}

        # Menentukan kodon x_i
        for i in range(len(kodons)):
            
            # Melakukan permutasi terhadap kodon x_i
            kodons_11, kodons_22 = permutasi_kodons(kodons[i])
            
            # Menentukan kodon x_j
            for j in range(len(kodons)):
                
                # Menentukan kodon x_k
                for k in range(len(kodons)):

                    # Memeriksa apakah kodon x_i, kodon x_j, dan kodon x_k ekuivalen siklik    
                    if kodons[j] == kodons_11 and kodons[k] == kodons_22:

                        # Memeriksa apakah kodon x_i sama dengan kodon x_j dan kodon x_k
                        if kodons[i] == kodons_11 and kodons[i] == kodons_22:
                            urutan_kodon[kodons[i]] = [kodons[i]]
                        else:
                            urutan_kodon[kodons[i]] = [kodons[i],kodons[j], kodons[k]]

                        # Menampilkan elemen ekuivalen siklik pada kodon x_i
                        print(f'Elemen Ekuivalen Siklik pada Kodon {kodons[i]}: {kodons[i], kodons_11, kodons_22}\n')
                    else:
                        continue

        # Mengembalikan list dari urutan kodon
        return list(urutan_kodon.values())

    # Menyimpan elemen-elemen yang ekuivalen siklik ke dalam file
    def simpan_ke_file(nama_file, data):
        with open(nama_file, 'w') as file:
            for i,urutan in enumerate(data):
                file.write(f'{i+1}. Elemen Ekuivalen Siklik pada Kodon {kodons[i]}: {urutan}\n\n')
    
    # Mendefinisikan urutan kodon
    urutan_kodon = elemen_ekuivalen_siklik(kodons)
    # Mendefinisikan nama file
    nama_file = 'Elemen-elemen yang Ekuivalen Siklik.txt'
    # Menyimpan elemen-elemen yang ekuivalen siklik ke dalam file
    simpan_ke_file(nama_file, urutan_kodon)
    # Mengembalikan urutan kodon
    return urutan_kodon

# Memanggil fungsi Mencari_Elemen_Ekuivalen_siklik
elemen_ekuivalen_siklik = Mencari_Elemen_Ekuivalen_siklik(kodons)     
```

# Kelas Ekuivalen Siklik

```python
# Membentuk fungsi untuk mencari kelas-kelas ekuivalen siklik
def KelasEkuivalenSiklik(urutan_kodon):

    # Mensortir dan mengelompokkan kelas-kelas ekuivalen siklik
    def KelompokKelasEkuivalenSiklik(strs):

        # Membuat dictionary untuk menyimpan kelas-kelas ekuivalen siklik
        Kelas = {}
        for word in strs:

            # Mengurutkan karakter pada kelas-kelas ekuivalen siklik
            key = "".join(sorted(word))
            
            # Memeriksa apakah key ada dalam Kelas-kelas ekuivalen siklik
            if key in Kelas or len(key) < 9:
                continue
            else:
                Kelas[key] = word
        
        # Kembalikan daftar kelas-kelas ekuivalen siklik
        return list(Kelas.values())
    
    def simpan_ke_file(nama_file, data):
        with open(nama_file, 'w') as file:
            for i,urutan in enumerate(data):
                file.write(f'{i+1}. Kelas ekuivalen [{urutan[0]}]: {urutan}\n\n')

    hasil = KelompokKelasEkuivalenSiklik(urutan_kodon)
    nama_file = 'Kelas Ekuivalen Siklik.txt'
    print("Kelas-kelas ekuivalen:")
    for i, kelompok in enumerate(hasil):
        print(f"{i+1}. Kelas ekuivalen [{kelompok[0]}]: {kelompok}")
    simpan_ke_file(nama_file, hasil)
    return hasil

Kelas_Ekuivalen = KelasEkuivalenSiklik(elemen_ekuivalen_siklik)
```
